```python
menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
price = [2000, 3000, 3000, 2500, 2500, 3000]
class Kiosk: 
    def __init__(self):
        self.menu = menu
        self.price = price
        self.order_menu = []  # 주문 리스트
        self.order_price = []  # 가격 리스트

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ' : ', self.price[i], '원')

    # 주문 메서드
    def menu_select(self): 
        print()  # 줄 바꿈
        n = 0
        while n < 1 or len(menu) < n:
            n = int(input("음료 번호를 입력하세요 : "))  # 음료 번호 입력

            # 메뉴판에 있는 음료 번호일때
            if 1 <= n & n <= len(menu):
                self.price_sum = self.price[n-1]  # 선택 음료의 가격
        
            # 메뉴판에 없는 번호일 때
            else:  
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 음료 온도 물어보기
        t = 0  # 기본값을 넣어주고
        while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
            t= int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))       
            if t == 1:
                self.temp = 'HOT'
            elif t == 2:
                self.temp = 'ICE'

            else:    
                print("1과 2 중 하나를 입력하세요.\n")
        self.order_price.append(self.price[n-1])
        self.order_menu.append(self.temp + '  ' + self.menu[n-1])
        print('주문 음료', self.temp, self.menu[n-1], ' : ', self.price[n-1], '원')  # 온도 속성을 추가한 주문 결과 출력

    #추가 주문
        while n != 0:
            print ()    # 줄 바꿈
            n = int(input("추가 주문은 음료 번호를, 지불은 0을 입력하세요  :  "))
            if n > 0 and n < len(self.menu) + 1:        
                
            # 추가 음료 온도
                t = 0
                while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
                    t= int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))       
                    if t == 1:
                        self.temp = 'HOT'
                    elif t == 2:
                        self.temp = 'ICE'

                    else:    
                        print("1과 2 중 하나를 입력하세요.\n")
                self.order_price.append(self.price[n-1])
                self.order_menu.append(self.temp + '  ' + self.menu[n-1])
                self.price_sum += self.price[n-1]

                print("추가 주문 음료", self.temp, self.menu[n-1], '  :  ' , self.price[n-1], '원\n', '합계  :  ', self.price_sum, '원')
            else :
                if n == 0 :  ## 지불 입력
                    print("주문이 완료되었습니다. ")
                    print(self.order_menu, self.order_price)
                else :    ##  없는 번호를 입력할 때
                    print("없는 메뉴입니다. ")

    # 지불 메서드
    def pay(self):
    
        print(f'합계 금액은 총 {self.price_sum} 입니다.')
    
        self.pay_method = 0
    
        while self.pay_method != 1 and self.pay_method != 2:
            self.pay_method = int(input("현금 지불은 1번, 카드 지불은 2번을 눌러주세요 : "))
            if self.pay_method == 1:
                print("직원을 호출하겠습니다.")
            elif self.pay_method == 2:
                print('IC방향에 맞게 카드를 꽂아주세요')
            else:
                print('다시 결제를 시도해주세요')

    
    def table(self):
    # 외곽
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

# 주문 상품명 : 해당 금액
        for i in range(len(self.order_menu)):
            print(self.order_menu[i], '  :  ' , self.order_price[i], '원\n')

        print('합계 금액 :', self.price_sum)

# 외곽
        for i in range(5):
            print('|' + ' ' * 31+ '|')
        print('⟝' + '-' * 30 + '⟞')


 


                    
    


            
```

```python
a = Kiosk()
a.menu_print()
a.menu_select()
a.pay()
a.table()
```

```python
## 사각형 넓이
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')
        
        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')
        
        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    def par(self):
        width, vertical = map(int, input('밑변, 높이를 입력하세요. 예시 : 밑변,높이\n >>>').split(','))  #가로, 세로를 입력받는 코드
        area = width*vertical# [[YOUR CODE]]  # 넓이를 구하는 부분
        result = '평행사변형의 넓이는 : ' + str(area) # [[YOUR CODE]]# [[YOUR CODE]]
        return result

    def trape(self):
        lower, upper, height = map(int, input('밑변,윗변, 높이를 입력하세요. 예시 : 밑변,윗변,높이\n >>>').split(','))  #가로, 세로를 입력받는 코드
        area = 0.5*(lower+upper)*height# [[YOUR CODE]]  # 넓이를 구하는 부분
        result = '사다리꼴의 넓이는 : ' + str(area) # [[YOUR CODE]]
        return result

a = Square()
```
