```python
# Q1-2

def is_palindrome(text):
    return text == text[::-1]
    
def palindrome_classifier(words):

    result = {"회문O": [], "회문X": []}

    count = 0

    
    for word in words:

        if is_palindrome(word):
            result["회문O"].append(word)
            count += 1

        else:
            result["회문X"].append(word)

    # 회문 비율 계산
    ratio = count / len(words)

    return result, count, ratio


targets = [
    '기러기',
    '까마귀',
    '토마토',
    '스윗소스',
    '스위스',
    '인도인',
    '한국인',
    '별똥별',
    '우영우',
    '우향우'
]

print(palindrome_classifier(targets))
```

```python
#Q2-1

scores = {
    "유원영": { "국어": 94,"영어": 91,"수학": 89 },

    "설지민": {"국어": 88,"영어": 96,"수학": 90},

    "신윤아": {"국어": 86,"영어": 88,"수학": 77 },

    "장윤아": {"국어": 98,"영어": 76,"수학": 92}
}
```

```python
# Q2-2
def score_analysis(name):

    student = scores[name]

    print(f"[{name} 성적 분석]")

    total_student = 0
    total_class = 0

    # 과목별 분석
    for subject in ["국어", "영어", "수학"]:

        student_score = student[subject]

        # 과목 평균 계산
        subject_avg = 0

        for s in scores:
            subject_avg += scores[s][subject]

        subject_avg /= len(scores)

        diff = student_score - subject_avg

        print(f"{subject}: {student_score}점, "
              f"학급 평균({subject_avg:.2f}점) 대비 "
              f"{diff:+.2f}")

        total_student += student_score
        total_class += subject_avg

    # 전체 평균 계산
    student_avg = total_student / 3
    class_avg = total_class / 3

    total_diff = student_avg - class_avg

    print()
    print(f"전과목 평균: {student_avg:.2f}점, "
          f"학급 평균({class_avg:.2f}점) 대비 "
          f"{total_diff:+.2f}")



score_analysis("설지민")
```
