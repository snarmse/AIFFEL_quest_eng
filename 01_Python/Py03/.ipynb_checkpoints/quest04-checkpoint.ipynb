```python
## math 모듈에서 제곱연산자 지원
from math import pow

## 정수일때 까지 입력 받는 함수 정의
def get_num(prompt):
    while True:    ## 정수일 때 까지 입력받으므로 반복문 형태로 작성
        try: 
            return int(input(prompt)) ## 정수값 입력받으면 반환하고 반복문 종료하는 return

        except ValueError:   
            print("잘못된 입력입니다. 정수를 입력해주세요.") ## 정수가 아닐 시 에러문 작성하고 계속 반복
    
##계산기 반복시행하는 반복문
while True:
    
    num1 = get_num("첫번째 정수를 입력하세요: ")
    num2 = get_num("두번째 정수를 입력하세요: ") ## 함수를 이용해 두 정수 입력
           
    ans = 0  ## 결과값 변수 설정
## 연산자 입력
    try:     
        oper = input("연산자를 입력하세요: ")    ##  연산자 입력
        
        if oper == '+':
                ans = num1 + num2
        elif oper == '-':
                ans = num1 - num2            
        elif oper == '*':
                ans = num1 * num2
        elif oper == '/':
                ans = num1 / num2           ## 각 연산자의 경우 계산 수행
        elif oper == '**':
                ans = int(pow(num1, num2))    ## power는 float반환하므로 int
        else:
            raise ValueError('지원하지 않는 연산자입니다!')
            
    except ZeroDivisionError:        ## 0으로 나눌 시 zerodivision error 출력
        print('0으로 나눌 수 없습니다!')
        
    except ValueError as e:        ##except문을 추가해서 코드가 raise에서 종료되지 않도록 한다.
        print(e)                    ## 에러문 출력

        
    else:
        print(f"{ans}")       ## 예외가 아닐 시 결과값 출력
    
    ##계산기 계속 사용하는지 여부

    calc = input('계속 계산하시겠습니까? (y, n)') ## 계산 여부 입력
    if calc != 'y':
        break                   ## 계산 계속 안 할시 반복문 종료

        
```

반복문, 예외처리, 조건문이 복합적으로 쓰이기 때문에 들여쓰기에 유의하며 코딩을 해야했다.
raise문을 쓰면 바로 error가 뜨고 프로그램이 종료되는데, 이걸 except ValueError문을 추가해서 프로그램을 계속 진행할 수 있다.
진현님은 반복문을 두개 쓰는 대신 정수 입력받는 함수를 설정했는데 코드가 더 간단해진다. 반복문이 두 개 일땐 구조가 헷갈릴 수 있으므로 코드가 어떻게 하면 간단해지는지 늘 고민해야겠다.
