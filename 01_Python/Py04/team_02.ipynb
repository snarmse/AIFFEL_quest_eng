Account 클래스를 생성한 후 생성자(hint: 매직매서드..!!)를 구현해보세요.  
생성자에서는 예금주와 초기 잔액만 입력 받습니다.   
은행이름은 SC은행으로 계좌번호는 3자리-2자리-6자리 형태로 랜덤하게 생성됩니다.   
(은행이름: SC은행, 계좌번호: 111-11-111111)  
* Q2. 클래스 변수: 클래스 변수를 사용해서 Account 클래스로부터 생성된 계좌 객체의 개수를 저장하세요.  
* Q3. 클래스 변수 출력 : Account 클래스로부터 생성된 계좌의 개수를 출력하는 get_account_num() 메서드를 추가하세요.  
* Q4. 입금 메서드 : Account 클래스에 입금을 위한 deposit 메서드를 추가하세요. 입금은 최소 1원 이상만 가능합니다.  
* Q5. 출금 메서드 : Account 클래스에 출금을 위한 withdraw 메서드를 추가하세요.  
출금은 계좌의 잔고 이상으로 출금할 수는 없습니다.  
* Q6. 정보 출력 메서드 : Account 인스턴스에 저장된 정보를 출력하는 display_info() 메서드를 추가하세요.  
잔고는 세자리마다 쉼표를 출력하세요.  
(은행이름: SC은행, 예금주: 파이썬, 계좌번호: 111-11-111111, 잔고: 10,000원)  
* Q7. 이자 지급하기 : 입금 횟수가 5회가 될 때 잔고를 기준으로 1%의 이자가 잔고에 추가되도록 코드를 변경해보세요.  
* Q8. 여러 객체 생성 : Account 클래스로부터 3개 이상 인스턴스를 생성하고 생성된 인스턴스를 리스트에 저장해보세요.  
* Q9. 객체 순회 반복문을 통해 리스트에 있는 객체를 순회하면서 잔고가 100만원 이상인 고객의 정보만 출력하세요.  
* Q10. 입금과 출금 내역이 기록되도록 코드를 업데이트 하세요.    
(입금 내역과 출금 내역을 출력하는 deposit_history와 withdraw_history 메서드를 추가하세요.)  


```python
import random

class Account:
    # 클래스 변수: 생성된 계좌 객체의 개수를 저장
    account_count = 0

    # 생성자(__init__) 구현
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        self.bank_name = 'SC은행'

        # 3자리-2자리-6자리 형태 랜덤 계좌번호 생성( 015- 이런 형식도 가능)
        self.account_number = f'{random.randint(1, 999):03}-{random.randint(0, 99):02}-{random.randint(0, 999999):06}'

        # 입금 카운트 변수
        self.deposit_count = 0

        # 입금/출금 내역 리스트
        self.deposit_log = []
        self.withdraw_log = []

        # 객체가 생성될 때마다 클래스 변수 1씩 증가
        Account.account_count += 1

    # 클래스 변수 출력 메서드
    def get_account_num(self):
        # 클래스명.클래스변수명 Account.account_count 형식으로 직접 접근
        print(f'현재 개설된 총 계좌 수: {Account.account_count}개')
        return Account.account_count

    # 입금 메서드
    def deposit(self, amount):
        if amount >= 1:
            self.balance += amount
            self.deposit_count += 1
            self.deposit_log.append(amount)  # 입금 내역 기록

            # 입금 5회가 될 때 1% 이자 지급
            if self.deposit_count % 5 == 0:
                interest = self.balance * 0.01
                self.balance += interest
                self.deposit_log.append(f'이자 1% 지급 (+{int(interest):,}원)') # 이자 내역도 기록
        else:
            print('입금은 최소 1원 이상만 가능합니다.')

    # 출금 메서드
    def withdraw(self, amount):
        if amount > self.balance:
            print('잔고가 부족하여 출금할 수 없습니다.')
        else:
            self.balance -= amount
            self.withdraw_log.append(amount)  # 출금 내역 기록

    # 정보 출력 메서드
    def display_info(self):
        # 잔고는 {:,} 포맷을 사용하여 세 자리마다 쉼표 출력
        print(f'은행이름: {self.bank_name}, 예금주: {self.name}, 계좌번호: {self.account_number}, 잔고: {int(self.balance):,}원')

    # 입금 내역 출력 메서드
    def deposit_history(self):
        print(f'--- {self.name} 고객님 입금 내역 ---')
        for log in self.deposit_log:
            # log가 문자열(이자 지급 내역)인 경우와 숫자(입금액)인 경우를 나누어 출력
            if type(log) == str:
                print(log)
            else:
                print(f'{log:,}원 입금') ## ,: 천단위 구분 기호

    # 출금 내역 출력 메서드
    def withdraw_history(self):
        print(f'--- {self.name} 고객님 출금 내역 ---')
        for log in self.withdraw_log:
            print(f'{log:,}원 출금')

```

```python
print("=== 실행 ===")
# 3개 이상 인스턴스 생성 후 리스트에 저장
account1 = Account("파이썬", 1500000)
account2 = Account("자바", 50000)
account3 = Account("씨플플", 2500000)

account_list = [account1, account2, account3]

account1.get_account_num()
print()

# 테스트를 위한 입출금 작업 진행
for i in range(6):
    account1.deposit(10000) # 5회차 입금 - 이자 발생 구간
account1.withdraw(50000)

account2.deposit(5000)
account2.withdraw(60000) # 잔고 부족 에러 확인

print()
print("=== 100만원 이상 고객 정보 출력 ===")
# 리스트 객체 순회하며 잔고 100만원 이상 고객 출력
for acc in account_list:
    if acc.balance >= 1000000:
        acc.display_info()

print()
print("=== 입금과 출금 내역 기록 확인 ===")
account1.deposit_history()
print()
account1.withdraw_history()
```

    * battle 함수

        * 만약 Player 인스턴스가 살아남았다면

        * Player 인스턴스가 살아남지 못했을 경우

* Q4. main 함수 만들기
    * 몬스터의 이름, 레벨이 매핑된 딕셔너리 정의하기
        * monster_dict = {‘슬라임’: 1, ‘고블린’: 2, ‘오크’: 3}
    * 사용자로부터 이름을 입력받아 Player 인스턴스 생성하기
    * 몬스터 딕셔너리로부터 Monster 인스턴스 생성하기
    * 생성된 Monster 인스턴스와 Player 인스턴스가 battle 함수를 통해 전투
        * player는 생성된 몬스터 3마리(슬라임, 고블린, 오크)와 모두 전투해야함
        * 전투 도중에 Player가 사망하면 이후 전투를 진행하지 않고 게임오버 출력하고 종료

```python
import random


# Character 클래스 (부모 클래스)
class Character:
    # 이름, 레벨, 체력, 공격력, 방어력 속성 초기화
    def __init__(self, name, level, hp, attack_power, defense):
        self.name = name
        self.level = level
        self.hp = hp
        self.attack_power = attack_power
        self.defense = defense

    # 체력이 0보다 큰지 확인하여 살았는지 죽었는지 bool(True/False) 반환
    def is_alive(self):
        if self.hp > 0:
            return True
        else:
            return False

    # 공격을 받았을 때 체력이 감소하는 메서드
    def take_damage(self, damage):
        # 실제 받는 데미지 = 적의 공격력 - 내 방어력
        actual_damage = damage - self.defense

        # 방어력이 더 높아서 데미지가 마이너스가 되면 0으로 처리 (체력 회복 방지)
        if actual_damage < 0:
            actual_damage = 0

        self.hp -= actual_damage
        return actual_damage # 입힌 데미지를 출력하기 위해 값을 반환

    # 타겟에게 데미지를 입히고 상황을 출력하는 메서드
    def attack_target(self, target):
        # 1부터 내 공격력 사이의 랜덤한 데미지 발생
        damage = random.randint(1, self.attack_power)

        # 타겟에게 데미지를 입히고 실제 들어간 데미지 값을 받아옴
        actual_damage = target.take_damage(damage)    ## monster.take_damage(damage)

        # 이미지의 출력 양식에 맞게 전투 텍스트 출력
        print(f"{self.name}이(가) {target.name}에게 {actual_damage} 만큼 공격했다...!")
        print(f"{target.name}의 체력: {target.hp}")


# Player 클래스 (자식 클래스)

class Player(Character):
    # 플레이어는 이름만 입력받고 나머지는 고정값으로 시작
    def __init__(self, name):
        # super().__init__()을 사용해 부모(Character)의 속성을 그대로 가져옴
        super().__init__(name, level=1, hp=100, attack_power=25, defense=5)
        self.experience = 0 # 플레이어 전용 속성: 경험치 추가

    def gain_experience(self, exp):
        self.experience += exp
        # 진행 상황을 알기 쉽게 경험치 획득 출력
        print(f"경험치를 {exp} 획득했습니다. (현재 경험치: {self.experience})")

    def level_up(self):
        if self.experience >= 50:
            self.level += 1
            self.attack_power += 10
            self.defense += 5
            self.experience -= 50 # 레벨업 후 남은 경험치 처리
            print(f"\n*** 레벨 업! {self.name}의 레벨이 {self.level}이(가) 되었습니다! ***\n")

# Monster 클래스 (자식 클래스)
class Monster(Character):
    def __init__(self, name, level):
        # 몬스터의 레벨에 비례하여 랜덤한 능력치 생성
        hp = random.randint(10, 30) * level
        attack_power = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level

        # 생성된 능력치로 부모 클래스의 초기화 메서드 실행
        super().__init__(name, level, hp, attack_power, defense)



# battle 함수 (전투 시스템)
def battle(player, monster):
    print(f"\n{monster.name} 과의 전투를 시작합니다.")

    # 둘 다 살아있는 동안 무한 반복 (자동 사냥)
    while player.is_alive() and monster.is_alive():
        # 플레이어가 먼저 몬스터를 공격
        player.attack_target(monster)

        # 몬스터가 살아있다면 플레이어를 반격
        if monster.is_alive():
            monster.attack_target(player)

    # 반복문이 끝남 -> 누군가 죽었다는 뜻
    if player.is_alive():
        print("전투 승리!")
        print(f"{monster.name} 을(를) 이겼다!")

        # 보상 지급 및 레벨업 체크
        exp = monster.level * 20
        player.gain_experience(exp)
        player.level_up()
        return True # 승리했으므로 True 반환

    else:
        print("전투 패배..")
        return False # 패배했으므로 False 반환



# main 함수 (게임 실행)

def main():
    # 몬스터 딕셔너리 준비
    monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3}

    print("=== 자동사냥 RPG에 오신 것을 환영합니다 ===")
    user_name = input("플레이어의 이름을 입력하세요: ")
    player = Player(user_name)

    # 딕셔너리의 항목들을 순회하며 몬스터 생성 및 전투
    for m_name, m_level in monster_dict.items():
        # 몬스터 인스턴스 생성
        monster = Monster(m_name, m_level)

        # 전투 시작 (반환값이 True면 승리, False면 패배)
        result = battle(player, monster)

        # 전투에서 졌을 경우 게임 오버 처리
        if result == False:
            print("\n게임오버")
            break # 더 이상 순회하지 않고 게임 종료


# 게임 실행
main()
```
